# **Computing technical indicators**

We are creating features as technical indicators that are set for different time-ranges from 5 to 50 days, with 5 days of difference. They are the following:

- SMA: Simple Moving Average.
- EMA: Exponential Moving Average.
- ATR: Average True Range.
- ADX: Average Directional Index.
- CCI: Commodity Channel Index.
- ROC: Rate of Change.
- RSI: Relative Strength Index.
- Force Index.
- MFI: Money Flow Index.
- OBV: On Balance Volume.
- UI: ULCER Index.

```python
f_names = []
for i in range(5, 51, 5):
    # SMA
    n = trend.sma_indicator.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] = trend.sma_indicator(df.close, i, fillna=False)
    # EMA
    n = trend.ema_indicator.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] = trend.ema_indicator(df.close, i, fillna=False)
    # ATR
    n = volatility.average_true_range.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] = volatility.average_true_range(df.high, df.low, df.close, i, fillna=False)
    # ADX
    n = trend.adx.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] = trend.adx(df.high, df.low, df.close, i, fillna=False)
    # CCI
    n = trend.cci.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] = trend.cci(df.high, df.low, df.close, i, fillna=False)
    # ROC
    n = momentum.roc.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] = momentum.roc(df.close, i, fillna=False)
    # RSI
    n = momentum.rsi.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] = momentum.rsi(df.close, i, fillna=False)
    # Force index
    n = volume.force_index.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] = volume.force_index(df.close, df.volume, i, fillna=False)
    # MFI
    n = volume.money_flow_index.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] = volume.money_flow_index(df.high, df.low, df.close, df.volume, i, fillna=False)
    # OBV
    n = volume.on_balance_volume.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] = volume.on_balance_volume(df.close, df.volume, i)
    # UI
    n =  volatility.ulcer_index.__name__ + str(i)
    print(n)
    f_names.append(n)
    df[n] =  volatility.ulcer_index(df.close, i, fillna=False)

# Add features to the dataset
f_names.extend([volume.acc_dist_index.__name__, others.cumulative_return.__name__, others.daily_log_return.__name__, others.daily_return.__name__])

df[volume.acc_dist_index.__name__] = volume.acc_dist_index(df.high, df.low, df.close, df.volume, fillna=False)

df[others.cumulative_return.__name__] =  others.cumulative_return(df.close, fillna=False)
  
df[others.daily_log_return.__name__] =  others.daily_log_return(df.close, fillna=False)
  
df[others.daily_return.__name__] =  others.daily_return(df.close, fillna=False)
```

# **Feature selection**

**In this section, we will introduce the three methods for feature selection.**

## **Filter methods**

### **Most Correlated features with the target**

```python
# Create the correlation matrix
corr = df.corr()[['close']]

# Chosing the features that are correlated (above 0.9),
# or negatively correlated with the target (below -0.9)
F1 = corr[abs(corr['close']) > 0.9].index.tolist()
F1
```

### **F-regression test**

```python
from sklearn.datasets import make_regression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

data = df.drop('close', axis = 1)
fs = SelectKBest(score_func = f_regression, k = 15)
fs_tr = fs.fit_transform(data, df.close)
F2 = data.columns[(fs.get_support())].tolist()
F2
```

## **Embedded methods**

### **RandomForestRegressor**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel

data = df.drop('close', axis = 1)
sel = SelectFromModel(RandomForestRegressor(n_estimators = 500))
sel.fit(data, df.close.values)
F3 = data.columns[(sel.get_support())].tolist()
F3
```

### **Lasso regularization**

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Lasso
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, df.close, test_size=0.2, shuffle=False)
```

```python
pipeline = Pipeline([
                     ('scaler', MinMaxScaler()),
                     ('model', Lasso())
])
```

```python
search = GridSearchCV(pipeline,
                      {'model__alpha':np.arange(0.1,10,0.1)}, 
                      scoring="neg_mean_squared_error",
                      cv=None
                      )

search.fit(X_train,y_train)
```

**The importance of a feature is the absolute value of its coefficient.**

```python
coefficients = search.best_estimator_.named_steps['model'].coef_
importance = np.abs(coefficients)
```

```python
F4 = np.array(df.columns)[importance > 0].tolist()
F4
```

## **Wrapper methods**

### **Bi-directional elimination (Step-wise Selection)**

```python
scaler = MinMaxScaler()
df_scaled =  df.copy()

df_scaled[df.columns.tolist()] = scaler.fit_transform(df_scaled[df.columns.tolist()])
```

```python
def stepwise_selection(data, target,SL_in=0.05,SL_out = 0.05):
    initial_features = data.columns.tolist()
    best_features = []

    while (len(initial_features) > 0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index = remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if (min_p_value < SL_in):
            best_features.append(new_pval.idxmin())
            while (len(best_features) > 0):
                best_features_with_constant = sm.add_constant(data[best_features])
                p_values = sm.OLS(target, best_features_with_constant).fit().pvalues[1:]
                max_p_value = p_values.max()
                if (max_p_value >= SL_out):
                    excluded_feature = p_values.idxmax()
                    best_features.remove(excluded_feature)
                else:
                    break 
        else:
            break
    return best_features
```

```python
data = df_scaled.drop('close', axis = 1)
```

```python
F5 = stepwise_selection(data, df_scaled.close)
F5
```
